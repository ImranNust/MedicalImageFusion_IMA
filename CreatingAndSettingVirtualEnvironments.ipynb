To check how many virtual environments are there, we use the following command:
```
mamba info --envs
```

---


To delete a virtual environment (For example, if there is an environment with the name 'test'), we use the following command:
```
mamba remove -n test --all
```

If the environment is created in a specific folder, then we will give the path, as follows:

```
mamba remove -p path_of_the_directory
```

** General Information **

On HiPerGator, Jupyter looks in two places for kernels when you launch a notebook:

1. /apps/jupyterhub/kernels/ for the globally available kernels that all users can use. (Also a good place to look for troubleshooting getting your own kernel going)
2. ~/.local/share/jupyter/kernels for each user. (Again, your home directory and the .local folder is hidden since it starts with a dot)

Make the ~/.local/share/jupyter/kernels directory: mkdir -p ~/.local/share/jupyter/kernels

Copy the /apps/jupyterhub/template_kernel folder into your ~/.local/share/jupyter/kernels directory:

cp -r /apps/jupyterhub/template_kernel/ ~/.local/share/jupyter/kernels/EnvironmentName
---

Step 4[Optional]: Create the directory with the name `kernels' in your home directory (~/.local/share/jupyter/) if it is not already created, as follows:

```
mkdir ~/.local/share/jupyter/kernels
```

Step 4b: Copy the `template-kernel` from the global path to your environement in the local path as follows:

```
cp -r /apps/jupyterhub/template_kernel/ ~/.local/share/jupyter/kernels/PyTorchForMedical/
```
**The name of template_kernel will be replaced with the name of your environment, which is 'PyTorchForMedical`**

---

Step 5: Now find out the files in the the PyTorchForMedical folder as follows:

```
ls ~/.local/share/jupyter/kernels/PyTorchForMedical/
```
we can see that there are four files as follows:
```
kernel.json  logo-32x32.png  logo-64x64.png  run.sh
```

---

Step 6: Open the kernel.json file and replace all content with the following content:

```
{
 "language": "python",
 "display_name": "PyTorchForMedicalEnv",
 "argv": [
  "~/.local/share/jupyter/kernels/PyTorchForMedical/run.sh",
  "-f",
  "{connection_file}"
 ]
}
```

**You can open json file using nano as follows:**

```
nano ~/.local/share/jupyter/kernels/PyTorchForMedical/kernel.json  
```
**To save, press ctrl+o and Enter. \\
To exit, press ctrl+x.**

---

Step 7: Open the run.sh file and replace all content with the following content:

```
#!/usr/bin/bash

exec /blue/weishao/share/conda/envs/PyTorchForMedical/bin/python -m ipykernel "$@"
```

**You can open sh file using nano as follows:**

```
nano ~/.local/share/jupyter/kernels/PyTorchForMedical/run.sh  
```
**To save, press ctrl+o and Enter. \\
To exit, press ctrl+x.**


# To install Packages

To install a package let's say numpy, use the following command
```
mamba install numpy==1.21.2
```

if you want to install the latest version of the numpy just remove the version number.

---

If you have a requirements file, and you want to install package from that file just write the following lines
```
mamba install --file requirements.txt -c conda-forge
```

"The last part '-c conda-forge' will tell the mamba to install from channel conda-forge"
